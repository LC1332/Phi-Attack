```python
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
torch.set_default_device('cuda')
```

```python


model = AutoModelForCausalLM.from_pretrained("microsoft/phi-1_5",
                                             trust_remote_code=True,
                                             torch_dtype="auto")

tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1_5",
                                          trust_remote_code=True,
                                          torch_dtype="auto")

```

```python
def generate(input_text, max_length=1000):
    inputs = tokenizer(input_text, return_tensors="pt", return_attention_mask=False)
    outputs = model.generate(**inputs, max_length=max_length)
    text = tokenizer.batch_decode(outputs)[0]
    return text
```

```python
def prepare_input(tokenizer, prompts):
    input_tokens = tokenizer.batch_encode_plus(prompts, return_tensors="pt", padding=True)
    for t in input_tokens:
        if torch.is_tensor(input_tokens[t]):
            input_tokens[t] = input_tokens[t].to('cuda')

    return input_tokens
```

```python
import json


def save_datas_into_file( datas, file_name ):
    output_file = file_name

    with open(output_file, 'w') as f:
        for data in datas:
            json.dump(data, f, ensure_ascii = False)
            f.write('\n')
```

```python
import os
import random
import string

def save_data_in_folder(datas, save_folder):
    # 检查保存文件夹是否存在，如果不存在则创建
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    for test_time in range(100):
        # 生成一个随机的文件名
        file_name = ''.join(random.choices(string.ascii_letters + string.digits, k=8)) + '.jsonl'
        file_path = os.path.join(save_folder, file_name)

        # 如果文件不存在，则保存数据并退出循环
        if not os.path.exists(file_path):
            save_datas_into_file(datas, file_path)
            return

    return
```

```python
def check_good( text ):
    if "\nSolution:" in text or "\nsolution:" in text or "\nAnswer:" in text or "\nanswer:" in text:
        return True
    return False
```

```python
batch_size = BATCH_SIZE
start_ids = [x for x in range(0, len(queries), batch_size)]

import random

random.shuffle(start_ids)
```

```python


# input_texts = queries[0:4]
batch_size = BATCH_SIZE

save_datas_fail = []
save_datas_good = []


first_save = False
first_save_size = 5

# save_datas = []

count = 0
count_good = 0

for i in tqdm(start_ids):
    tokenizer.add_special_tokens({'pad_token': ' '})
    tokenizer.padding_side = 'left'

    input_texts = queries[i:min(i+batch_size, len(queries))]

    encode_inputs = prepare_input(tokenizer, input_texts)

    outputs = model.generate(**encode_inputs, max_new_tokens= MAX_LEN )

    del tokenizer.special_tokens_map['pad_token']

    texts_from_batch = tokenizer.batch_decode(outputs, skip_special_tokens=False)

    batch_data = []
    for j in range(len(input_texts)):
        data = {
            'input': input_texts[j],
            'output': texts_from_batch[j]
        }
        if check_good(data['output']):
            save_datas_good.append(data)
            count_good += 1
        else:
            save_datas_fail.append(data)

        if not first_save and len(save_datas_good) >= first_save_size:
            save_data_in_folder(save_datas_good, good_save_folder)
            save_datas_good = []
            save_data_in_folder(save_datas_fail, bad_save_folder)
            save_datas_fail = []
            first_save = True
            print("First Save Done")


        if len(save_datas_good) >= 1000:
            save_data_in_folder(save_datas_good, good_save_folder)
            save_datas_good = []

        if len(save_datas_fail) >= 1000:
            save_data_in_folder(save_datas_fail, bad_save_folder)
            save_datas_fail = []

    count += 1

    if count % 5 == 0:
        print( count_good ,  " / ", count * batch_size)



    # save_datas.extend(batch_data)

```
